```python
s_cctv["최근증가율"] = (s_cctv["2014년"]+s_cctv["2015년"]+s_cctv["2016년"]) / s_cctv["2013년도 이전"] * 100
```

```python
pop_s.head()
```

```python
pop_s.drop([0],inplace=True)
```

```python
pop_s["외국인비율"] = (pop_s["외국인"]/pop_s["인구수"]) * 100
pop_s["고령자비율"] = (pop_s["고령자"]/pop_s["인구수"]) * 100
```

```python
pop_s.head()
```

```python
pop_s.sort_values(by="인구수",ascending=False).head()
```

```python
len(s_cctv), len(pop_s)
```

```python
d_res = pd.merge(s_cctv,pop_s,on="구별")
```

```python
d_res.head()
```

```python
unused_year_columns = ["2013년도 이전","2014년","2015년","2016년"]
d_res.drop(unused_year_columns, axis=1, inplace=True)
```

```python
d_res.head()
```

```python
d_res.set_index(keys="구별",inplace=True)
```

```python
d_res.head()
```

```python
d_res["CCTV비율"] = (d_res["소계"]/d_res["인구수"])*100
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = d_res["인구수"].values.reshape(-1,1)
y = d_res["소계"]
```

```python
y
```

```python
model = LinearRegression()
```

```python
#x.values
```

```python
#x.values.reshape(-1,1)
```

```python
model.fit(x,y)
```

```python
len(d_res)
```

```python
fx = np.linspace(100000,700000,25)
fx
```

```python
fx = fx.reshape(-1,1)
```

```python
predicted = model.predict(fx)
```

```python
len(fx), len(predicted)
```

```python
fx.shape
```

```python
predicted.shape
```

```python
fx_flattend = fx.flatten()
predict_flattend = predicted.flatten()
```

```python
fx_flattend.shape, predict_flattend.shape
```

```python
len(fx_flattend), len(predict_flattend)
```

```python
model_df = pd.DataFrame({
    "인구수":fx_flattend,
    "예측 CCTV수":predict_flattend
})
```

```python
model_df.head()
```

```python
#scatter: 점그래프(숫자,숫자)
#plot:선그래프(숫자,숫자)

fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(d_res["인구수"],d_res["소계"])
ax.plot(model_df["인구수"],model_df["예측 CCTV수"],color="red")

plt.title("인구수 대비 CCTV 갯수")
plt.xlabel("인구수")
plt.ylabel("CCTV")

ax.grid()
plt.show()
```

##9.강조하고 싶은 데이터 시각화

- 필요한 재료 2가지
- 1:실제값(실제 설치된 cctv)
- 2:예측값(예측된 cctv)
- 실제값-예측값=오차

```python
d_res.head(1)
```

```python
d_res["소계"] #첫번째
```

```python
model_df["예측 CCTV수"]
model.predict(x) #두번째
```

```python
d_res["소계"] - model.predict(x)
```

```python
d_res["오차"] = d_res["소계"] - model.predict(x)
```

```python
d_res.head()
```

```python
df_sort_f = d_res.sort_values(by="오차",ascending=False)
df_sort_t = d_res.sort_values(by="오차")
```

```python
d_res["인구수"][0], d_res["소계"][0], d_res.index[0]
```

```python
from matplotlib.colors import ListedColormap

color_step = ["#FF0000","#00FF00","#0000FF","#FF00FF","#FFFF00","#00FFFF"]
my_map=ListedColormap(color_step)
```

```python
fig, ax = plt.subplots(figsize=(14,10))
result = ax.scatter(d_res["인구수"], d_res["소계"], c=d_res['오차'], cmap=my_map)
ax.plot(model_df['인구수'],model_df['예측 CCTV수'],color="red")

#plt.text(df_sort_f["인구수"][10], df_sort_f["소계"][10], df_sort_f.index[10])

for i in range(5):
  plt.text(df_sort_t["인구수"][i] * 1.02, df_sort_t["소계"][i] * 0.98, df_sort_t.index[i], fontsize=10)

for i in range(5):
  plt.text(df_sort_f["인구수"][i], df_sort_f["소계"][i], df_sort_f.index[i], fontsize=10)

ax.set(xlabel="인구수",ylabel="CCTV",title="인구수 대비 CCTV 갯수")

plt.colorbar(result)
plt.grid()
plt.show()
```
