```python
import tensorflow as tf
import numpy as np
tf.random.set_seed(5)
```

```python
arr = np.loadtxt("data-03-diabetes.csv",delimiter=',',dtype=np.float32)
print(arr.shape)
#학습 데이터 분리: 70%, 531개
#검증 데이터 분리: 30%, 228개
x_train = arr[:531,:-1]
x_test = arr[531:,:-1]
y_train = arr[:531,[-1]]
y_test = arr[531:,[-1]]
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
W=tf.Variable(tf.random.normal([8,1]),name='weight')
b=tf.Variable(tf.random.normal([1]),name='bias')
print(W,b)
```

```python
#H(X) = sigmoid(W*X + b)
def hy(X):
    return tf.sigmoid(tf.matmul(X,W) + b) # value is between 0 to 1
```

```python
#logloss, 2진분류 모델
def cost_func():
    cost = -tf.reduce_mean(y_train*tf.math.log(hy(x_train)) +
                         (1-y_train)*tf.math.log(1-hy(x_train)))
    return cost
```

```python
#경사하강법 learning_rate(학습률)을 0.01로 설정하여 optimizer object 생성
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
print("===Start===")
for step in range(10000):
    opt.minimize(cost_func,var_list=[W,b])
    if step%1000 == 0:
        print('%d'%step,'cost:',cost_func().numpy())
print("===finished===")
```

```python
preds = tf.cast(hy(x_test) > 0.5,dtype=tf.float32)
tf.cast(tf.equal(y_test,preds),dtype=tf.float32)
```

```python
# 정확도 측정 : accuracy computation
def predict(X):
    return tf.cast(hy(X) > 0.5,dtype=tf.float32)

# 검증 데이터를 사용하여 예측 (70%:30%)
preds = predict(x_test)
accuracy = tf.reduce_mean(tf.cast(tf.equal(preds,y_test),dtype=tf.float32))
print(tf.reduce_sum(tf.cast(tf.equal(preds,y_test),dtype=tf.float32)))
print('Accuracy:',accuracy.numpy()) # Accuracy: 0.78070176
# print('Hypothesis:\n',hypothesis(x_test).numpy())
# print('Predict:\n',preds.numpy())
```
