## 2. 서울시 인구 데이터 불러오기

```python
pop_data = "/content/Seoul_Population.xls"
pop_seoul = pd.read_excel(pop_data, header=2, usecols="B, D, G, J, N")
```

```python
pop_seoul.head()
```

```python
pop_seoul.columns
```

```python
pop_seoul.columns[2]
```

```python
pop_seoul.rename(columns={
    pop_seoul.columns[0]: "구별",
    pop_seoul.columns[1]: "인구수",
    pop_seoul.columns[2]: "한국인",
    pop_seoul.columns[3]: "외국인",
    pop_seoul.columns[4]: "고령자",
}, inplace=True)
```

## 3. 데이터 내림차순, 오름차순 정렬해보기

## 4. 기존 데이터를 활용한 새로운 컬럼 만들기

```python
seoul_cctv["최근증가율"] = ((seoul_cctv["2014년"] + seoul_cctv["2015년"] + seoul_cctv["2016년"]) / seoul_cctv["2013년도 이전"]) * 100
```

```python
# seoul_cctv.구별
```

```python
# 최근증가율이 높은 상위 5개 구
seoul_cctv.sort_values(by="최근증가율", ascending=False).head()
```

## 5. 서울시 인구현황 데이터 훑어보기

```python
pop_seoul.head()
```

```python
# "합계" 행 삭제
pop_seoul.drop([0], inplace=True)
```

```python
pop_seoul.head(2)
```

```python
pop_seoul["외국인비율"] = (pop_seoul["외국인"] / pop_seoul["인구수"]) * 100
pop_seoul["고령자비율"] = (pop_seoul["고령자"] / pop_seoul["인구수"]) * 100
```

```python
pop_seoul.head()
```

```python
# 인구수가 많은 곳 상위 10개 구
pop_seoul.sort_values(by="인구수", ascending=False).head(10)
```

## 6. 서울시 CCTV 데이터와 서울시 인구수 데이터 병합

```python
seoul_cctv.head(1)
```

```python
pop_seoul.head(1)
```

```python
# 조건1. 데이터의 갯수가 같아야 한다
len(seoul_cctv), len(pop_seoul)
```

```python
# 조건2. 기준 컬럼
# "구별" 이라는 공통된 값을 가진 컬럼을 이용
```

```python
data_result = pd.merge(seoul_cctv, pop_seoul, on="구별")
```

```python
data_result.head()
```

```python
# 컬럼 삭제
# axis 축
# axis=0 가로축, 행
# axis=1 세로축, 열
unused_year_columns = ["2013년도 이전", "2014년", "2015년", "2016년"]
data_result.drop(unused_year_columns, axis=1, inplace=True)
```

```python
data_result.head(2)
```

## 7. 구별 컬럼을 인덱스로 설정

```python
data_result.set_index(keys="구별", inplace=True)
```

```python
data_result.head()
```

```python
data_result.index
```

```python
data_result["CCTV비율"] = (data_result["소계"] / data_result["인구수"]) * 100
```

```python
data_result.head()
```

```python
data_result.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# 정답 데이터, y
# 정답을 도출하기 위한 데이터, x
# 인구수가 몇 일때, 적절한 CCTV 갯수
x = data_result["인구수"].values.reshape(-1, 1)
y = data_result["소계"]
```

```python
# 선형회귀모델 객체 생성
model = LinearRegression()
```

```python
# 선형회귀모델 학습
model.fit(x, y)
```

```python
len(data_result)
```

```python
data_result.head(2)
```

```python
fx = np.linspace(100000, 700000, 25)
len(fx)
```

```python
fx = fx.reshape(-1, 1)
```

```python
# 특정 인구수마다 몇개의 CCTV가 있는 것이 적당한지 판단하는 모델
predicted = model.predict(fx)
```

```python
len(fx), len(predicted)
```

```python
fx.shape
```

```python
predicted.shape
```

```python
fx_flattened = fx.flatten()
```

```python
predict_flattened = predicted.flatten()
```

```python
fx_flattened.shape, predict_flattened.shape
```

```python
model_df = pd.DataFrame({
    "인구수": fx_flattened,
    "예측 CCTV 수": predict_flattened
})
```

```python
model_df.head(2)
```

```python
# scatter: 점그래프(숫자, 숫자)
# plot: 선그래프(숫자, 숫자)

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(data_result["인구수"], data_result["소계"])
ax.plot(model_df["인구수"], model_df["예측 CCTV 수"], color="red")

ax.grid()
ax.set(xlabel='인구수', ylabel='CCTV',
       title='인구수 대비 CCTV')

plt.show()
```

## 9. 강조하고 싶은 데이터 시각화

```python
# 필요한 재료가 2가지
# 첫 번째: 실제값 (실제 설치된 CCTV 갯수)
# 두 번째: 예측값 (모델이 예측한 CCTV 갯수)
# 실제값 - 예측값 = 오차
# 강남구: 실제값(3000) - 예측값(1500) = 1500
```

```python
# 첫 번째
data_result["소계"]
```

```python
# 두 번째
model.predict(x)
```

```python
# 실제값 - 예측값
data_result["소계"] - model.predict(x)
```

```python
# 오차 컬럼을 데이터프레임 추가
data_result["오차"] = data_result["소계"] - model.predict(x)
```

```python
data_result.head()
```

```python
df_sort_f = data_result.sort_values(by="오차", ascending=False)  # 내림차순
```

```python
df_sort_t = data_result.sort_values(by="오차", ascending=True)  # 오름차순
```

```python
df_sort_t.head()
```

```python
data_result["인구수"][0], data_result["소계"][0], data_result.index[0]
```

```python
for i in range(5):
  print(df_sort_f["인구수"][i], df_sort_f["소계"][i], df_sort_f.index[i])
```

```python
from matplotlib.colors import ListedColormap

color_step = ["#f00a0a", "#f0880a", "#d1f00a",
              "#12f00a", "#0af0d9", "#6e0af0"]

my_map = ListedColormap(color_step)
```

```python
fig, ax = plt.subplots(figsize=(14, 10))
result = ax.scatter(data_result["인구수"], data_result["소계"],
           c=data_result["오차"], cmap=my_cmap)
ax.plot(model_df["인구수"], model_df["예측 CCTV 수"], color="red")

# plt.text() 함수는 그래프상에 텍스트를 추가하는 데 사용
# 첫 번째 매개변수는 텍스트를 표시할 x 좌표
# 두 번째 매개변수는 텍스트를 표시할 y 좌표
# 세 번째 매개변수는 그래프상에 표시할 텍스트 내용
# 적절한 CCTV가 설치된 곳
plt.text(
    df_sort_f["인구수"][10],
    df_sort_f["소계"][10],
    df_sort_f.index[10],
    fontsize=12
    )

# CCTV가 많은 곳 상위 5개 이름을 그래프에 표현
for i in range(5):
  plt.text(
      df_sort_f["인구수"][i] * 1.02,
      df_sort_f["소계"][i] * 0.98,
      df_sort_f.index[i],
      fontsize=12
      )

# CCTV가 적은 곳 상위 5개 이름을 그래프에 표현
for i in range(5):
  plt.text(
      df_sort_t["인구수"][i] * 1.02,
      df_sort_t["소계"][i] * 0.98,
      df_sort_t.index[i],
      fontsize=12
      )

ax.set(xlabel="인구수", ylabel="CCTV", title="인구수 대비 CCTV 설치 현황")

plt.colorbar(result)
plt.grid()
plt.show()
```
